```python
from PIL import Image
from pylab import *

im = array(Image.open('images/profile.jpg').convert('L'))
im
```

```python
im2 = 255 - im # negative
im3 = (100.0/255) * im + 100 # clamp to 100..200
im4 = 255.0 * (im / 255.0) ** 2
gray()
imshow(im2)
show()
imshow(im3)
show()
imshow(im4)
show()
```
