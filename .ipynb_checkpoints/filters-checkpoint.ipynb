```python
from PIL import Image, ImageFilter
from pylab import *

im0 = Image.open('images/moon.jpg')
figure(figsize=(15, 15))

gray()

subplot(3, 4, 1)
plt.imshow(im0)
plt.title('original')

subplot(3, 4, 2)
im2 = im0.filter(ImageFilter.CONTOUR)
plt.imshow(im2)
plt.title('Contour')

subplot(3, 4, 3)
im3 = im0.filter(ImageFilter.DETAIL)
plt.imshow(im3)
plt.title('Detail')

subplot(3, 4, 4)
im4 = im0.filter(ImageFilter.EDGE_ENHANCE)
plt.imshow(im4)
plt.title('Edge Enhance')

subplot(3, 4, 5)
im5 = im0.filter(ImageFilter.EDGE_ENHANCE_MORE)
plt.imshow(im5)
plt.title('Edge Enhance More')

subplot(3, 4, 6)
im6 = im0.filter(ImageFilter.EMBOSS)
plt.imshow(im6)
plt.title('Emboss')

subplot(3, 4, 7)
im7 = im0.filter(ImageFilter.FIND_EDGES)
plt.imshow(im7)
plt.title('Find Edges')

subplot(3, 4, 8)
im8 = im0.filter(ImageFilter.SMOOTH)
plt.show(im8)
plt.title('Lowpass 1')

subplot(3, 4, 9)
im9 = im0.filter(ImageFilter.SMOOTH_MORE)
plt.imshow(im9)
plt.title('Lowpass 2')

subplot(3, 4, 10)
im10 = im0.filter(ImageFilter.SHARPEN)
plt.imshow(im10)
plt.title('Sharpen')

# custom

size = (3, 3)
kernel = [1, 1, 1, 0, 0, 0, -1, -1, -1]
kern1 = ImageFilter.Kernel(size, kernel, scale=None, offset=0)
subplot(3, 4, 11)
im11 = im0.filter(kern1)
plt.imshow(im11)
plt.title('Custom 1')

kernel = [1, 0, -1, 1, 0, -1, 0, 0, -1]
kern2 = ImageFilter.Kernel(size, kernel, scale=None, offset=0)
subplot(3, 4, 12)
im12 = im0.filter(kern2)
plt.imshow(im12)
plt.title('Custom 2')

plt.show()
```

```python
# mean filter
import scipy.ndimage
import numpy as np
from PIL import Image

a = Image.open('images/profile.jpg').convert('L')

k = np.ones((5, 5)) / 25
b = scipy.ndimage.filters.convolve(a, k)
b = Image.fromarray(b)
b
```

```python
# min filter
a = Image.open('images/lena_noisy.png')
b = scipy.ndimage.filters.minimum_filter(a, size=5, footprint=None, output=None, mode='reflect', cval=0.0, origin=0)
b = Image.fromarray(b)
b
```
