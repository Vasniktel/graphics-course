```python
# prewitt
import cv2
from skimage import img_as_ubyte, img_as_int
from scipy import ndimage
import numpy as np
from matplotlib import pyplot as plt

moon = cv2.imread('images/moon.jpg', 0)
img = img_as_int(moon)

prewitt_vertical = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
], dtype='float64')

prewitt_vertical_out = img_as_ubyte(ndimage.convolve(img, prewitt_vertical, mode='constant', cval=0.0))
plt.imshow(prewitt_vertical_out, cmap='gray')
plt.show()
```

```python
prewitt_horizontal = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1]
], dtype='float64')

prewitt_horizontal_out = img_as_ubyte(ndimage.convolve(img, prewitt_horizontal, mode='constant', cval=0.0))
plt.imshow(prewitt_horizontal_out, cmap='gray')
plt.show()
```

```python
# sobel
from PIL import Image

moon = cv2.imread('images/moon.jpg', 0)
img = img_as_int(moon)

sobel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype='float64')

sobel_out = img_as_ubyte(ndimage.convolve(img, sobel, mode='constant', cval=0.0))
Image.fromarray(sobel_out)
```

```python
import cv2
import numpy as np

img = cv2.imread('images/profile.jpg', 0)

sobel_x = cv2.Sobel(img, -1, 1, 0, ksize=5)
sobel_y = cv2.Sobel(img, -1, 0, 1, ksize=5)

plt.imshow(sobel_x, cmap='gray')
plt.show()
plt.imshow(sobel_y, cmap='gray')
plt.show()
```

```python
# laplacian
img = cv2.imread('images/profile.jpg')
output = cv2.Laplacian(img, -1)
plt.imshow(output)
plt.show()
```
